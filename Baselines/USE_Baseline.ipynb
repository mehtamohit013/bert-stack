#Sentence Encoder

```
from absl import logging

import tensorflow as tf

import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import re
import seaborn as sns

module_url = "https://tfhub.dev/google/universal-sentence-encoder/4"
model = hub.load(module_url)
print ("module %s loaded" % module_url)
def embed(input):
  return model(input)
```

#Neural Network

```
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras import regularizers

model_nn = Sequential()
model_nn.add(Dense(128, input_dim=1024, activation='relu'))
model_nn.add(Dense(32, activation='relu'))
model_nn.add(Dense(16,activation='relu'))
model_nn.add(Dense(8,activation='relu'))
model_nn.add(Dense(1, activation='sigmoid'))
```

#Evaluating Perfomance on Android data

```
from google.colab import drive
drive.mount('/gdrive')
```

```
import pickle
with open('/gdrive/MyDrive/sbert_supervised/data_android.txt','rb') as d:
    data = pickle.load(d)
with open('/gdrive/MyDrive/sbert_supervised/score_android.txt','rb') as s:
    score = pickle.load(s)
```

```
data_emb = list()
for i in range(0,len(data)):
    temp = embed(data[i])
    a = list()
    for _ , temp_emb in enumerate(np.array(temp).tolist()):
        a.append(temp_emb)
    data_emb.append(a[0]+a[1])
    if(i%1000==0):
        print(f'calucalted {i} embeddings')
print(len(data_emb[0]))

```

```
with open ('gdrive/MyDrive/sbert_supervised/USE_android.txt','wb') as a:
    pickle.dump(data_emb,a)
```

```
with open ('gdrive/MyDrive/sbert_supervised/USE_android.txt','rb') as a:
    data_emb = pickle.load(a)
```

```
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

print(len(data_emb),len(score))
data_emb,score = shuffle(data_emb,score)
data_emb_train,data_emb_dev,score_train,score_dev = train_test_split(data_emb,score,test_size = 0.1, random_state=13)
```

```
model_nn.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
model_nn.fit(data_emb_train,score_train,epochs = 20, batch_size=16)
_, accuracy = model_nn.evaluate(data_emb_dev, score_dev)
print('Accuracy: %.2f' % (accuracy*100))
```
