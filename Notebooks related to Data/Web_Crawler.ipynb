#Final Draft

```
with open('/content/android_data_title.json') as data1:
  questions = pd.read_json(data1,orient='table')
cnt  = 0 
for id in list(questions.index.values):
    
    if(questions.loc[id,'Related'] == None):
        url = "https://android.stackexchange.com/questions/"+str(id)
        page = requests.get(url)
        soup = BeautifulSoup(page.text,'lxml')
        start = time.time()
        try:
            related = soup.find('div',class_ = 'related js-gps-related-questions')
            for related_q in related.find_all('div',class_='spacer'):
                related_id = related_q.a['href'].split('/')
                if(questions.at[id,'Related']==None):
                    questions.at[id,'Related'] = []
                questions.at[id,'Related'].append(int(related_id[2]))
            end = time.time()
            cnt+=1
            print(cnt)
            if(end-start<5):
                time.sleep(1-(end-start))
        except Exception  as e :
            print(soup)
            print("No related value found for id : "+str(id))
            print(e)
            print(url)
```
