```
def text_refine(text):
    text = text.replace('\n',' ')
    text = text.replace('\r',' ')
    text = text.replace('\\','')
    return text
```

```
import json
import pandas as pd
import numpy as np
```

```
corpus = ''
```

```
with open('/content/drive/My Drive/Data/android/questions_related.json') as q:
    questions_adr = pd.read_json(q,orient='table')
questions_adr.info()
for i in questions_adr.index.values:
    text = questions_adr.at[i,'Title']
    # text = text_refine(text)
    corpus = corpus +' ' + text
    text = questions_adr.at[i,'Text']
    # text = text_refine(text)
    corpus = corpus +' '+text
len(corpus)
```
